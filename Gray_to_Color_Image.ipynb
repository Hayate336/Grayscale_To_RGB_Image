```python
font = cv2.FONT_HERSHEY_SIMPLEX
font_size = 1.1
font_thickness = 4
x,y = 50,900
```

```python
image = cv2.imread('/content/jjk.jpg')
b, g, r    = image[:, :, 0], image[:, :, 1], image[:, :, 2] # For RGB image
IM = cv2.merge((b,g,r))

print('Original Image Array --> ')
print(image)

print(' ')
print(' ')
print('Extracted Red Array --> ')
print(r)

print(' ')
print(' ')
print('Extracted Blue Array --> ')
print(g)

print(' ')
print(' ')
print('Extracted Green Array --> ')
print(b)
```

```python
image1 = cv2.imread('/content/jjk.jpg')
gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
(thresh, bw) = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

Oim = cv2.putText(image, 'Original Image', (x,y), font, font_size, (0,0,0), font_thickness, cv2.LINE_AA)
Gim = cv2.putText(gray, 'Grayscale Image', (x,y), font, font_size, (0,0,0), font_thickness, cv2.LINE_AA)
Bim = cv2.putText(bw, 'Binary Image', (x,y), font, font_size, (0,0,0), font_thickness, cv2.LINE_AA)
Nim = cv2.putText(IM, 'Grayscale To RGB Image', (x,y), font, font_size, (0,0,0), font_thickness, cv2.LINE_AA)

H1 = np.concatenate((Gim, Bim), axis=1)
cv2_imshow(H1)
H2 = np.concatenate((Oim, Nim), axis=1)
cv2_imshow(H2)
```
