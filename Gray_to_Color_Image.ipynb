```python
font = cv2.FONT_HERSHEY_SIMPLEX
font_size = 1.1
font_thickness = 4
x,y = 50,900
```

```python
image = cv2.imread('/content/jjk.jpg')
b, g, r    = image[:, :, 0], image[:, :, 1], image[:, :, 2] # For RGB image

print('Original Image Array --> ')
print(' ')
print(image)

print(' ')
print(' ')
print(' ')
print('Extracted Red Values Array --> ')
print(' ')
print(r)

print(' ')
print(' ')
print(' ')
print('Extracted Blue Values Array --> ')
print(' ')
print(g)

print(' ')
print(' ')
print(' ')
print('Extracted Green Values Array --> ')
print(' ')
print(b)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
IM = cv2.merge((gray,gray,gray))
Mimg = IM.copy()

print('Grayscale Image Array --> ')
print(' ')
print(gray)

print(' ')
print(' ')
print(' ')
print('Merged RGB Image Array --> ')
print(' ')
print(Mimg)

IM[:, :, 0], IM[:, :, 1], IM[:, :, 2] = b,g,r

print(' ')
print(' ')
print(' ')
print('Masked Merged RGB Image Array --> ')
print(' ')
print(IM)
```

```python
Oim = cv2.putText(image, 'Original Image', (x,y), font, font_size, (0,0,0), font_thickness, cv2.LINE_AA)
Gim = cv2.putText(gray, 'Grayscale Image', (x,y), font, font_size, (0,0,0), font_thickness, cv2.LINE_AA)
Mim = cv2.putText(Mimg, 'Merged Image', (x,y), font, font_size, (0,0,0), font_thickness, cv2.LINE_AA)
Nim = cv2.putText(IM, 'Grayscale To RGB Image', (x,y), font, font_size, (0,0,0), font_thickness, cv2.LINE_AA)

cv2_imshow(Gim)
H1 = np.concatenate((Oim, Mim, Nim), axis=1)
cv2_imshow(H1)
```
